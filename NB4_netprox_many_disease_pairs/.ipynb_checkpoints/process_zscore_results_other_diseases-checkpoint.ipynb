```python
# save the num overlap and overlap p-val in dataframes
network_num_overlap = pd.DataFrame(np.zeros((5,5)),index=['CHD','ID','EPI','TS','SCZ'])
network_num_overlap.columns = ['CHD','ID','EPI','TS','SCZ']

network_pval_overlap = pd.DataFrame(np.ones((5,5)),index=['CHD','ID','EPI','TS','SCZ'])
network_pval_overlap.columns = ['CHD','ID','EPI','TS','SCZ']


# how much does observed overlap compare to expected?
from scipy.stats import hypergeom
M = len(Gint.nodes())
# ----- CHD + ID -------
n = len(CHD_z_genes)
N = len(ID_z_genes)
x = len(np.intersect1d(CHD_z_genes,ID_z_genes))
p_temp = hypergeom.sf(x,M,n,N)
print('CHD + ID')
print(p_temp)

network_num_overlap.loc['CHD']['ID']=x
network_num_overlap.loc['ID']['CHD']=x

network_pval_overlap.loc['CHD']['ID']=p_temp
network_pval_overlap.loc['ID']['CHD']=p_temp

# ----- CHD + EPI -------
n = len(CHD_z_genes)
N = len(EPI_z_genes)
x = len(np.intersect1d(CHD_z_genes,EPI_z_genes))
p_temp = hypergeom.sf(x,M,n,N)
print('CHD + EPI')
print(p_temp)

network_num_overlap.loc['CHD']['EPI']=x
network_num_overlap.loc['EPI']['CHD']=x

network_pval_overlap.loc['CHD']['EPI']=p_temp
network_pval_overlap.loc['EPI']['CHD']=p_temp


# ----- CHD + TS -------
n = len(CHD_z_genes)
N = len(TS_z_genes)
x = len(np.intersect1d(CHD_z_genes,TS_z_genes))
p_temp = hypergeom.sf(x,M,n,N)
print('CHD + TS')
print(p_temp)

network_num_overlap.loc['CHD']['TS']=x
network_num_overlap.loc['TS']['CHD']=x

network_pval_overlap.loc['CHD']['TS']=p_temp
network_pval_overlap.loc['TS']['CHD']=p_temp

# ----- CHD + SCZ -------
n = len(CHD_z_genes)
N = len(SCZ_z_genes)
x = len(np.intersect1d(CHD_z_genes,SCZ_z_genes))
p_temp = hypergeom.sf(x,M,n,N)
print('CHD + SCZ')
print(p_temp)

network_num_overlap.loc['CHD']['SCZ']=x
network_num_overlap.loc['SCZ']['CHD']=x

network_pval_overlap.loc['CHD']['SCZ']=p_temp
network_pval_overlap.loc['SCZ']['CHD']=p_temp

# ----- ID + EPI -------
n = len(ID_z_genes)
N = len(EPI_z_genes)
x = len(np.intersect1d(ID_z_genes,EPI_z_genes))
p_temp = hypergeom.sf(x,M,n,N)
print('ID + EPI')
print(p_temp)

network_num_overlap.loc['ID']['EPI']=x
network_num_overlap.loc['EPI']['ID']=x

network_pval_overlap.loc['ID']['EPI']=p_temp
network_pval_overlap.loc['EPI']['ID']=p_temp


# ----- ID + TS -------
n = len(ID_z_genes)
N = len(TS_z_genes)
x = len(np.intersect1d(ID_z_genes,TS_z_genes))
p_temp = hypergeom.sf(x,M,n,N)
print('ID + TS')
print(p_temp)

network_num_overlap.loc['ID']['TS']=x
network_num_overlap.loc['TS']['ID']=x

network_pval_overlap.loc['ID']['TS']=p_temp
network_pval_overlap.loc['TS']['ID']=p_temp

# ----- ID + SCZ -------
n = len(ID_z_genes)
N = len(SCZ_z_genes)
x = len(np.intersect1d(ID_z_genes,SCZ_z_genes))
p_temp = hypergeom.sf(x,M,n,N)
print('ID + SCZ')
print(p_temp)

network_num_overlap.loc['ID']['SCZ']=x
network_num_overlap.loc['SCZ']['ID']=x

network_pval_overlap.loc['ID']['SCZ']=p_temp
network_pval_overlap.loc['SCZ']['ID']=p_temp


# ----- EPI + TS -------
n = len(EPI_z_genes)
N = len(TS_z_genes)
x = len(np.intersect1d(EPI_z_genes,TS_z_genes))
p_temp = hypergeom.sf(x,M,n,N)
print('EPI + TS')
print(p_temp)

network_num_overlap.loc['EPI']['TS']=x
network_num_overlap.loc['TS']['EPI']=x

network_pval_overlap.loc['EPI']['TS']=p_temp
network_pval_overlap.loc['TS']['EPI']=p_temp

# ----- EPI + SCZ -------
n = len(EPI_z_genes)
N = len(SCZ_z_genes)
x = len(np.intersect1d(EPI_z_genes,SCZ_z_genes))
p_temp = hypergeom.sf(x,M,n,N)
print('EPI + SCZ')
print(p_temp)

network_num_overlap.loc['EPI']['SCZ']=x
network_num_overlap.loc['SCZ']['EPI']=x

network_pval_overlap.loc['EPI']['SCZ']=p_temp
network_pval_overlap.loc['SCZ']['EPI']=p_temp

# ----- TS + SCZ -------
n = len(TS_z_genes)
N = len(SCZ_z_genes)
x = len(np.intersect1d(TS_z_genes,SCZ_z_genes))
p_temp = hypergeom.sf(x,M,n,N)
print('TS + SCZ')
print(p_temp)

network_num_overlap.loc['TS']['SCZ']=x
network_num_overlap.loc['SCZ']['TS']=x

network_pval_overlap.loc['EPI']['TS']=p_temp
network_pval_overlap.loc['TS']['EPI']=p_temp


```

```python
# make sure significant overlaps have >0 overlapping genes
network_pval_overlap = np.minimum(network_pval_overlap*(network_num_overlap>0)+(network_pval_overlap+1)*(network_num_overlap==0),1)

display(network_num_overlap)
display(network_pval_overlap)

# network_num_overlap.to_csv('network_num_overlap_z'+str(zthresh)+'.tsv',sep='\t')
# network_pval_overlap.to_csv('network_pval_overlap_z'+str(zthresh)+'.tsv',sep='\t')
```

```python
if d1=='CHD': 
    d1_DNV=CHD_DNV
elif d1=='ID':
    d1_DNV=ID_DNV
elif d1=='EPI':
    d1_DNV=EPI_DNV
elif d1=='TS':
    d1_DNV=TS_DNV
elif d1=='SCZ':
    d1_DNV=SCZ_DNV
    
if d2=='CHD': 
    d2_DNV=CHD_DNV
elif d2=='ID':
    d2_DNV=ID_DNV
elif d2=='EPI':
    d2_DNV=EPI_DNV
elif d2=='TS':
    d2_DNV=TS_DNV
elif d2=='SCZ':
    d2_DNV=SCZ_DNV
    
print(len(d1_DNV))
print(len(d2_DNV))
    
```

```python
# make a table to store the results of DNV enrichment
DNV_enrichment_df = pd.DataFrame(np.zeros((1,6)))
DNV_enrichment_df.columns=[d1+'_DNVs_num_overlap',d1+'_DNV_pval_overlap',
                          d2+'_DNVs_num_overlap',d2+'_DNV_pval_overlap',
                          d1+'_'+d2+'_shared_DNVs_num_overlap',
                          d1+'_'+d2+'_shared_DNVs_pval_overlap']

# ------ d1 overlap -----------
M = len(Gint)
n = len(np.setdiff1d(d1_DNV,d1_seeds)) # remove seeds
N = len(high_z_genes)
x = len(np.intersect1d(high_z_genes,d1_DNV))
p_temp = hypergeom.sf(x,M,n,N)
print(d1+' DNVs overlapping with network intersection')
print(x)
print(p_temp)

DNV_enrichment_df[d1+'_DNVs_num_overlap']=x
DNV_enrichment_df[d1+'_DNVs_pval_overlap']=p_temp

# ------ d2 overlap -----------
M = len(Gint)
n = len(np.setdiff1d(d2_DNV,d2_seeds)) # remove seeds
N = len(high_z_genes)
x = len(np.intersect1d(high_z_genes,d2_DNV))
p_temp = hypergeom.sf(x,M,n,N)
print('\n'+d2+' DNVs overlapping with network intersection')
print(x)
print(p_temp)

DNV_enrichment_df[d2+'_DNVs_num_overlap']=x
DNV_enrichment_df[d2+'_DNVs_pval_overlap']=p_temp

# ------ d1-d2 shared overlap -----------
M = len(Gint)
n = len(np.setdiff1d(np.intersect1d(d1_DNV,d2_DNV),d1_seeds+d2_seeds)) # remove seeds
N = len(high_z_genes)
x = len(set.intersection(set(d1_DNV),set(d2_DNV),set(high_z_genes)))
p_temp = hypergeom.sf(x,M,n,N)
print('\n'+d1+'-'+d2+' shared DNVs overlapping with network intersection')
print(x)
print(p_temp)

DNV_enrichment_df[d1+'_'+d2+'_shared_DNVs_num_overlap']=x
DNV_enrichment_df[d1+'_'+d2+'_shared_DNVs_pval_overlap']=p_temp


# DNV_enrichment_df.to_csv('DNV_enrichment_'+d1+'_'+d2)
DNV_enrichment_df
```

# Network localization of DNVs

```python
# import network_bio_toolkit for localization methods

sys.path.append('/Users/brin/network_bio_toolkit/network_bio_toolkit')

import Heat2
```

```python

heat = Heat2.Heat(gene_type = 'symbol', species = 'human')
heat.DG_universe=Gint
```

```python
# set numreps to 1000 for full run

heat.DEG_list = CHD_DNV
heat.DEG_list = list(np.intersect1d(heat.DEG_list,heat.DG_universe.nodes()))
print(len(heat.DEG_list))

num_reps=100
CHD_numedges_list, CHD_numedges_rand, CHD_LCC_list, CHD_LCC_size_rand = heat.localization(num_reps = num_reps, 
                                                                     sample_frac = 0.8, 
                                                                     method = 'both', 
                                                                     plot = False,print_counter=True)
```

```python
plt.figure(figsize=(4,3))
sns.set_style('white')
sns.kdeplot(CHD_numedges_list,color='r',lw=2,alpha=.6,shade=True,label='CHD DNVs')
sns.kdeplot(CHD_numedges_rand,color='k',lw=2,alpha=.5,shade=True,label='random')
plt.legend(loc='lower left',fontsize=12)
plt.ylabel('frequency',fontsize=16)
plt.xlabel('number edges in subgraph',fontsize=16)

# print the z-score and fdr
CHD_numedges_z = (np.mean(CHD_numedges_list)-np.mean(CHD_numedges_rand))/np.std(CHD_numedges_rand)

from scipy.special import ndtr
print(1-ndtr(CHD_numedges_z))

from scipy.stats import ks_2samp
print(ks_2samp(CHD_numedges_list,CHD_numedges_rand))

plt.savefig('../figures/CHD_DNVs_numEdges_localization.png',dpi=300,bbox_inches='tight')
```

# Pathway analysis of individual disease networks

```python
from gprofiler import GProfiler
gp = GProfiler("MyToolName/0.1")
```

```python
import os

savedir_profile = 'gprofile_cluster_'+focal_d + '_z'+str(zthresh)
if not os.path.isdir(savedir_profile):
    os.mkdir(savedir_profile)
    
# annotate the full subgraph
gp_results = pd.DataFrame(gp.gprofile(G_sub.nodes(),correction_method=gp.THR_FDR))
        
if len(gp_results)>0: # only save if there are any significant annotations
    gp_results.columns = ["query.number", "significant", "p.value", "term.size",
                          "query.size", "overlap.size", "recall", "precision",
                          "term.id", "domain", "subgraph.number", "term.name",
                          "relative.depth", "intersection"]
    print(gp_results[['p.value','term.id','term.name']].head())

    writer = pd.ExcelWriter(savedir_profile+'/full_subgraph_'+str(focal_cluster)+'.xlsx')
    gp_results.to_excel(writer)
    writer.save()


cluster2top_pathway = pd.DataFrame(columns=['cluster','pathway','term_ID','p-value','num_genes_in_cluster'])
for focal_cluster in partition.value_counts().index.tolist():
    print(focal_cluster)
    focal_genes = partition[partition==focal_cluster].index.tolist()
    if len(focal_genes)>10:
        gp_results = pd.DataFrame(gp.gprofile(focal_genes,custom_bg = partition.index.tolist(),correction_method=gp.THR_FDR))
        
        if len(gp_results)>0: # only save if there are any significant annotations
            gp_results.columns = ["query.number", "significant", "p.value", "term.size",
                                  "query.size", "overlap.size", "recall", "precision",
                                  "term.id", "domain", "subgraph.number", "term.name",
                                  "relative.depth", "intersection"]
            print(gp_results[['p.value','term.id','term.name']].head())
            cluster2top_pathway = cluster2top_pathway.append(pd.DataFrame({'cluster':[focal_cluster],
                                                                          'pathway':gp_results.loc[0]['term.name'],
                                                                           'term_ID':gp_results.loc[0]['term.id'],
                                                                          'p-value':gp_results.loc[0]['p.value'],
                                                                          'num_genes_in_cluster':[len(focal_genes)]}))

            writer = pd.ExcelWriter(savedir_profile+'/cluster_'+str(focal_cluster)+'.xlsx')
            gp_results.to_excel(writer)
            writer.save()

#gp_results = pd.DataFrame(gp.gprofile(focal_genes,correction_method=gp.THR_FDR))
```
